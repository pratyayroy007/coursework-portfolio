# Pandas Series Basics

Nine short exercises on creating, filtering, and comparing `pandas.Series` objects, plus set-style operations (intersection, union, symmetric difference) between them.

## Q1 — Create a Series from a list, a NumPy array, and a dict

```python

import pandas as pd
import numpy as np

s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series(np.array([50, 60, 70, 80]))
s3 = pd.Series({'A': 100, 'B': 200, 'C': 300})
print(s1)
print(s2)
print(s3)
```

## Q2 — Items in s1 that are NOT present in s2

```python

s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([4, 5, 6, 7, 8])
print(s1[~s1.isin(s2)])
```

## Q3 — Intersection of two Series

```python

print(pd.Series(np.intersect1d(s1, s2)))
```

## Q4 — Union of two Series

```python

print(pd.Series(np.union1d(s1, s2)))
```

## Q5 — Symmetric difference of two Series

```python

print(pd.Series(np.setxor1d(s1, s2)))
```

## Q6 — Count of each unique value in a Series

```python

s = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
print(s.value_counts())
```

## Q7 — Positions of elements divisible by 3

```python

s = pd.Series([1, 3, 5, 6, 9, 10, 12, 15])
print(np.where(s % 3 == 0)[0])
```

## Q8 — Extract items at given positional indices

```python

s = pd.Series([10, 20, 30, 40, 50])
print(s.take([0, 2, 4]))
```

## Q9 — Find the positional index in s2 of each item in s1

```python

s1 = pd.Series([2, 4, 6, 8])
s2 = pd.Series([1, 2, 3, 4, 5, 6, 7, 8])
print([s2[s2 == i].index[0] for i in s1])
```
