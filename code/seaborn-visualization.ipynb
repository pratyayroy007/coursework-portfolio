# Seaborn Visualization Basics

Exploring the built-in Titanic dataset with Seaborn's core plot types: histograms, bar/scatter/violin plots, pair plots, facet grids, and correlation heatmaps.

## Setup — load the Titanic dataset

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# List all built-in Seaborn datasets
print(sns.get_dataset_names())

titanic = sns.load_dataset('titanic')
print(titanic.head())
```

## Histogram — distribution of passenger age

```python

sns.histplot(data=titanic, x='age')
plt.title('Histogram of Age')
plt.show()
```

## Bar plot — average fare by passenger class

```python

sns.barplot(data=titanic, x='pclass', y='fare')
plt.title('Bar Plot')
plt.show()
```

## Scatter plot — age vs. fare

```python

sns.scatterplot(data=titanic, x='age', y='fare')
plt.title('Scatter Plot')
plt.show()
```

## Violin plot — fare distribution

```python

sns.violinplot(data=titanic, y='fare')
plt.title('Violin Plot')
plt.show()
```

## Pair plot — pairwise relationships across all numeric columns

```python

sns.pairplot(titanic)
plt.show()
```

## Facet grid — fare histograms split by class and sex

```python

g = sns.FacetGrid(titanic, col='pclass', row='sex')
g.map_dataframe(sns.histplot, x='fare')
plt.show()
```

## Heatmap — correlation between numeric features

```python

corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

## Cluster map — correlation heatmap with hierarchical clustering

```python

corr = titanic.corr(numeric_only=True)
sns.clustermap(corr, annot=True)
plt.show()
```

## Joint plot — age vs. fare with marginal distributions

```python

sns.jointplot(data=titanic, x='age', y='fare')
plt.show()
```
