# Categorical Encoding Techniques

Ten exercises covering label encoding, ordinal encoding, and one-hot encoding with scikit-learn, including inverse-transform and combined encoding pipelines.

## Setup

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

## Question 1 — Label encoding + inverse transform

```python

data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Result': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass']
}
df = pd.DataFrame(data)
le = LabelEncoder()
df['Encoded_Result'] = le.fit_transform(df['Result'])
print(df)

print("Mapping:")
for i, label in enumerate(le.classes_):
    print(label, "=", i)

decoded = le.inverse_transform(df['Encoded_Result'])
print("Decoded Labels:")
print(decoded)
```

## Question 2 — Ordinal encoding with an explicit category order

```python

data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Grade': ['Poor', 'Excellent', 'Average', 'Good', 'Poor']
}
df = pd.DataFrame(data)
order = [['Poor', 'Average', 'Good', 'Excellent']]
oe = OrdinalEncoder(categories=order)
df['Encoded_Grade'] = oe.fit_transform(df[['Grade']])
print(df)
```

## Question 3 — One-hot encoding a nominal category

```python

data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Branch': ['CSE', 'IT', 'ECE', 'CSE', 'IT']
}
df = pd.DataFrame(data)
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['Branch']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())
print(encoded_df)
```

## Question 4 — Ordinal encoding for education level

```python

data = {
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor']
}
df = pd.DataFrame(data)
order = [['High School', 'Bachelor', 'Master', 'PhD']]
oe = OrdinalEncoder(categories=order)
df['Encoded_Education'] = oe.fit_transform(df[['Education']])
print(df)
```

## Question 5 — Label encoding vs. one-hot encoding side by side

```python

data = {
    'Customer': ['A', 'B', 'C', 'D', 'E'],
    'City': ['Kolkata', 'Delhi', 'Mumbai', 'Chennai', 'Delhi']
}
df = pd.DataFrame(data)

# Label Encoding
le = LabelEncoder()
df['Label_Encoded'] = le.fit_transform(df['City'])
print(df)

# One-Hot Encoding
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['City']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())
print(encoded_df)
```

## Question 6 — One-hot encoding with generated feature count

```python

data = {
    'Car': ['Swift', 'Nexon', 'Tiago', 'Alto', 'Punch'],
    'Fuel': ['Petrol', 'Diesel', 'CNG', 'Petrol', 'Electric']
}
df = pd.DataFrame(data)
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['Fuel']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())
print(encoded_df)
print("Generated Features =", encoded_df.shape[1])
```

## Question 7 — A DataFrame with multiple categorical columns

```python

data = {
    'Name': ['Ram', 'Sita', 'John', 'Amit', 'Neha'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female'],
    'Experience': ['Junior', 'Senior', 'Mid', 'Senior', 'Junior']
}
df = pd.DataFrame(data)
print(df)
```

## Question 8 — One-hot encoding blood group categories

```python

data = {
    'Patient': ['A', 'B', 'C', 'D', 'E'],
    'Blood Group': ['A+', 'B+', 'O+', 'AB+', 'A-']
}
df = pd.DataFrame(data)
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['Blood Group']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())
print(encoded_df)
```

## Question 9 — Ordinal encoding for membership tiers

```python

data = {
    'Customer': ['A', 'B', 'C', 'D', 'E'],
    'Membership': ['Silver', 'Gold', 'Platinum', 'Gold', 'Silver']
}
df = pd.DataFrame(data)
order = [['Silver', 'Gold', 'Platinum']]
oe = OrdinalEncoder(categories=order)
df['Encoded'] = oe.fit_transform(df[['Membership']])
print(df)
```

## Question 10 — Combining label, one-hot, and ordinal encoding in one DataFrame

```python

data = {
    'Name': ['Ram', 'Rita', 'Amit', 'John', 'Neha'],
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female'],
    'City': ['Delhi', 'Kolkata', 'Delhi', 'Mumbai', 'Kolkata'],
    'Education': ['Bachelor', 'Master', 'PhD', 'High School', 'Bachelor'],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

# Gender -> label encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# City -> one-hot encoding
ohe = OneHotEncoder(sparse_output=False)
city = ohe.fit_transform(df[['City']])
city_df = pd.DataFrame(city, columns=ohe.get_feature_names_out())

# Education -> ordinal encoding
order = [['High School', 'Bachelor', 'Master', 'PhD']]
oe = OrdinalEncoder(categories=order)
df['Education'] = oe.fit_transform(df[['Education']])

final_df = pd.concat([df, city_df], axis=1)
print(final_df)
```
