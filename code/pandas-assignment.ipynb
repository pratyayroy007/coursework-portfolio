# Pandas Assignment — DataFrame Basics

Four exercises building DataFrames from dicts, nested dicts, and NumPy arrays, plus label- and boolean-based row selection.

## Question 1 — Build a DataFrame from a dict with a custom index

```python

import pandas as pd

data = {
    'Name': ['Anuj Maheta', 'Sandeep Oza', 'Manvi Sharma'],
    'Tickets': [5, 2, 6],
    'Amount': [1355, 1169, 1988]
}
index = ['T0001', 'T0002', 'T0003']
booking = pd.DataFrame(data, index=index)
print(booking)
```

## Question 2 — DataFrame with duplicate index labels, then filter/select

```python

data = {
    'Material': ['Wooden', 'Plywood', 'Plastic',
                 'Stainless Steel', 'Wooden', 'Aluminium'],
    'Colour': ['Maroon', 'Yellow', 'Red',
               'Silver', 'Light Blue', 'Golden'],
    'Price': [25000, 20000, 1500, 55000, 2500, 65000]
}
index = ['Sofa', 'Dining Table', 'Chair', 'Sofa', 'Chair', 'Dining Table']
furniture = pd.DataFrame(data, index=index)
print(furniture)

# Select rows by label (returns all matches for duplicate labels)
print(furniture.loc[['Chair', 'Sofa']])

# Boolean filtering
print(furniture[furniture['Price'] > 25000])
print(furniture[furniture['Price'] < 10000])

# Every other row
print(furniture.iloc[::2])
```

## Question 3 — DataFrame from a nested dict (outer keys become columns)

```python

data = {
    'Activity1': {'Blue House': 98, 'Red House': 87, 'Green House': 59, 'Yellow House': 78},
    'Activity2': {'Blue House': 85, 'Red House': 76, 'Green House': 67, 'Yellow House': 99},
    'Activity3': {'Blue House': 88, 'Red House': 80, 'Green House': 91, 'Yellow House': 55}
}
df = pd.DataFrame(data)
print(df)
```

## Question 4 — DataFrame from a NumPy array with custom index/columns

```python

import numpy as np

data = np.array([
    [98, 56, 87],
    [32, 76, 65],
    [55, 99, 88]
])
df = pd.DataFrame(data, index=[1, 2, 3], columns=['A', 'B', 'C'])
print(df)
```
