# Census & Life-Expectancy Data Analysis

Exploring a US county-level census dataset and a country-level GDP/life expectancy dataset with pandas grouping/filtering and matplotlib charts.

## Setup

Requires `census.csv` (US Census county-level data, as used in Coursera's *Applied Data Science with Python*) and `life.csv` (a gapminder-style country GDP/life-expectancy dataset) in the same directory as this notebook.

```

import pandas as pd
import matplotlib.pyplot as plt

census = pd.read_csv("census.csv")
census.describe()
```

## Assignment 1 — Which state has the most counties?

```

county_count = census.groupby('STNAME').count()['COUNTY']
print(county_count.idxmax())  # idxmax gives the STNAME with the most counties
```

## Assignment 2 — Three most common values of `CENSUS2010POP`

```

print(census['CENSUS2010POP'].value_counts().head(3))
```

## Assignment 3 — Counties with 2015 births between 14,000 and 16,000

```

result = census[(census['BIRTHS2015'] >= 14000) & (census['BIRTHS2015'] <= 16000)]
print(result[['STNAME', 'CTYNAME', 'BIRTHS2015']])
```

## Assignment 3b — Which states appear in that result?

```

states = result['STNAME'].drop_duplicates()
print(states)
```

## Assignment 4 — Top 10 countries by GDP per capita

```

life = pd.read_csv("life.csv")
top10 = life.sort_values(by='gdpPercap', ascending=False).head(10)
print(top10[['country', 'gdpPercap']])
```

## Assignment 5 — Bar chart of GDP per capita for the first 5 rows

```

data = life.head(5)
plt.figure(figsize=(6, 4))
plt.bar(data['country'], data['gdpPercap'])
plt.title('Bar Graph')
plt.xlabel('Country')
plt.ylabel('GDP per Capita')
plt.show()
```

## Assignment 6 — Pie chart of GDP per capita share

```

plt.figure(figsize=(6, 6))
plt.pie(data['gdpPercap'], labels=data['country'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

## Assignment 7 — GDP per capita vs. life expectancy scatter plot

```

plt.figure(figsize=(6, 4))
plt.scatter(data['gdpPercap'], data['lifeExp'])
plt.title('Scatter Plot')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

## Assignment 8 — Side-by-side line and bar subplots

```

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(data['country'], data['gdpPercap'])
ax[0].set_title('Line Graph')
ax[0].set_xlabel('Country')
ax[0].set_ylabel('GDP per Capita')

ax[1].bar(data['country'], data['gdpPercap'])
ax[1].set_title('Bar Graph')
ax[1].set_xlabel('Country')
ax[1].set_ylabel('GDP per Capita')

plt.tight_layout()
plt.show()
```
